#### Now let's work with the "south belt" of madrid. Make a subsetting that contains the following populations (level5): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆) 

```python
sur = ds[ds['level5'].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]
sur
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
def normalizar(df):
    return (df - df.min()) / (df.max() - df.min())

zonas = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
for zona in zonas:
    barrio = sur.groupby('level5').get_group(zona)
    barrio = normalizar(barrio['price'])
    barrio.hist(label=zona)

plt.legend()
plt.xlabel('Precio normalizado')
plt.ylabel('Frecuencia');
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
from ipyleaflet import Map, Marker, AwesomeIcon, basemaps

surmapa = sur.copy()
surmapa['latitude'] = surmapa['latitude'].str.replace(",", ".").astype(float)
surmapa['longitude'] = surmapa['longitude'].str.replace(",", ".").astype(float)

map = Map(center = (40.32, -3.78), zoom = 12, basemap=basemaps.Stamen.Terrain)

colores = {'Fuenlabrada': 'blue', 'Getafe': 'red', 'Leganés': 'orange', 'Alcorcón': 'gray'}

for barrio, grupo in surmapa.groupby('level5'):
    for indice, linea in grupo.iterrows():
        icono = AwesomeIcon(marker_color=colores[barrio],spin=False, )
        marker = Marker(location=(linea['latitude'], linea['longitude']), title=linea['level5'], icon=icono, draggable=False)
        map.add_layer(marker)

map

```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
getafemetrocuadrado = sur.groupby("level5").get_group("Getafe")
getafemetrocuadrado = getafemetrocuadrado.query('surface > 0')
getafemetrocuadrado = getafemetrocuadrado['price'] / getafemetrocuadrado['surface']
alcorconmetrocuadrado = sur.groupby("level5").get_group("Alcorcón")
alcorconmetrocuadrado = alcorconmetrocuadrado.query('surface > 0')
alcorconmetrocuadrado = alcorconmetrocuadrado['price'] / alcorconmetrocuadrado['surface']
print("El precio del metro cuadro en Alcorcón es mayor que el de Getafe por" , alcorconmetrocuadrado.mean() - getafemetrocuadrado.mean() , "mas")
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

#### What happens if we do the same analysis for price/m2 (★★☆) 

```python
galapagarmetrocuadrado = ds.groupby("level5").get_group("Galapagar")
galapagarmetrocuadrado = galapagarmetrocuadrado.query('surface > 0')
galapagarmetrocuadrado = galapagarmetrocuadrado['price'] / galapagarmetrocuadrado['surface']
valdemorillometrocuadrado = ds.groupby("level5").get_group("Valdemorillo")
valdemorillometrocuadrado = valdemorillometrocuadrado.query('surface > 0')
valdemorillometrocuadrado = valdemorillometrocuadrado['price'] / valdemorillometrocuadrado['surface']
print("La diferencia de precios por metro cuadrado entre Galapagar y Valdemorillo es", galapagarmetrocuadrado.mean() -  valdemorillometrocuadrado.mean())
```

#### Could you say there is a relation between surface and price? (★★☆) 
Hint: make a scatter plot.

```python
galapagar = ds.groupby('level5').get_group('Galapagar')
valdemorillo = ds.groupby('level5').get_group('Valdemorillo')

plt.scatter(valdemorillo['surface'], valdemorillo['price'], label='Valdemorillo')
plt.scatter(galapagar['surface'], galapagar['price'], label='Galapagar')

plt.xlabel('Superficie')
plt.gca().ticklabel_format(axis='y', style='plain')
plt.ylabel('Precio')
plt.legend()
plt.show()
```

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
fuenlabrada = ds.groupby('level5').get_group('Fuenlabrada')
getafe = ds.groupby('level5').get_group('Getafe')
leganes = ds.groupby('level5').get_group('Leganés')
alcorcon = ds.groupby('level5').get_group('Alcorcón')

plt.subplots(2, 2, figsize=(8,6))

plt.subplot(2,2,1)
plt.scatter(fuenlabrada['surface'], fuenlabrada['price'])
plt.xlabel('Superficie')
plt.gca().ticklabel_format(axis='y', style='plain')
plt.ylabel('Precio')
plt.title("Fuenlabrada")

plt.subplot(2,2,2)
plt.scatter(getafe['surface'],getafe['price'], color="red")
plt.xlabel('Superficie')
plt.gca().ticklabel_format(axis='y', style='plain')
plt.ylabel('Precio')
plt.title("Getafe")

plt.subplot(2,2,3)
plt.scatter(leganes['surface'],leganes['price'], color = "orange")
plt.xlabel('Superficie')
plt.gca().ticklabel_format(axis='y', style='plain')
plt.ylabel('Precio')
plt.title("Leganés")


plt.subplot(2,2,4)
plt.scatter(alcorcon['surface'],alcorcon['price'], color="grey")
plt.xlabel('Superficie')
plt.gca().ticklabel_format(axis='y', style='plain')
plt.ylabel('Precio')
plt.title("Alcorcón")

plt.tight_layout()
plt.show()
```
